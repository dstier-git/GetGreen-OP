# PART 1: Creating a Database and Table
What is SQL?
SQL (Structured Query Language) is a language for managing data in databases.
Think of a database as an organized collection of tables, like spreadsheets.

We'll create a simple database about students and their grades.

**Ignore the syntax for now. Try your best to read it like English!**

```python
# Create a connection to a new database (it will be created automatically)
conn = sqlite3.connect('school.db')
cursor = conn.cursor()

# Create a table called 'students'
cursor.execute('''
CREATE TABLE IF NOT EXISTS students (
    student_id INTEGER PRIMARY KEY,
    name TEXT NOT NULL,
    age INTEGER,
    grade_level INTEGER,
    gpa REAL
)
''');
```

Explanation of the table structure:
- student_id: A unique number for each student (PRIMARY KEY means it's unique)
- name: Student's name (TEXT means it's text/string data)
- age: Student's age (INTEGER means whole numbers)
- grade_level: What grade they're in (9, 10, 11, or 12)
- gpa: Grade Point Average (REAL means decimal numbers)

```python
# Insert some sample data
students_data = [
    (1, 'Alice Johnson', 16, 10, 3.8),
    (2, 'Bob Smith', 17, 11, 3.2),
    (3, 'Charlie Brown', 15, 9, 3.9),
    (4, 'Diana Prince', 18, 12, 3.7),
    (5, 'Eve Wilson', 16, 10, 3.5),
    (6, 'Frank Miller', 17, 11, 2.9),
    (7, 'Grace Lee', 15, 9, 4.0),
    (8, 'Henry Davis', 18, 12, 3.3)
]

cursor.executemany('INSERT OR IGNORE INTO students VALUES (?,?,?,?,?)', students_data)
conn.commit()

print("Database and table created with sample data!")
```

# PART 2: Retrieving Data

Moving forward, ignore all syntax except **the text after "query = "** and **before "result."** The rest is Python wrapping so that the SQL code ("SELECT" and such) can run inside this Python notebook.

`SELECT` is the most common SQL command. It retrieves data from tables.
Basic syntax: `SELECT` columns `FROM` table

```python
# Example 1: Select ALL columns from the students table
print("\n--- Example 1: SELECT * (all columns) ---")
query = "SELECT * FROM students"
result = pd.read_sql_query(query, conn)
print(result)

"""
The * means "all columns"
This shows every piece of information about every student
""";
```

```python
# Example 2: Select SPECIFIC columns
print("\n--- Example 2: SELECT specific columns ---")
query = "SELECT name, age, gpa FROM students"
result = pd.read_sql_query(query, conn)
print(result)

"""
Here we only get name, age, and gpa columns.
This is useful when you don't need all the information.
""";
```

```python
# Example 3: Select with calculated values
print("\n--- Example 3: SELECT with calculations ---")
query = "SELECT name, gpa, gpa * 25 AS percentage FROM students"
result = pd.read_sql_query(query, conn)
print(result)

"""
You can do math in SQL! 
AS creates an alias (nickname) for the calculated column.
Here we're converting GPA to a percentage.
""";
```

# PART 3: WHERE - Filtering Data
`WHERE` lets you filter results based on conditions.
Syntax: `SELECT` columns `FROM` table `WHERE` condition

```python
# Example 4: Filter by a specific value
print("\n--- Example 4: WHERE with equality ---")
query = "SELECT name, grade_level FROM students WHERE grade_level = 10"
result = pd.read_sql_query(query, conn)
print(result)

"""
This shows only students in grade 10.
The = sign checks for equality.
""";
```

```python
# Example 5: Filter with comparison operators
print("\n--- Example 5: WHERE with comparison ---")
query = "SELECT name, gpa FROM students WHERE gpa > 3.5"
result = pd.read_sql_query(query, conn)
print(result)
```

Common comparison operators:
- `>` greater than
- `<` less than
- `>=` greater than or equal to
- `<=` less than or equal to
- `=` equal to
- `!=` or `<>` not equal to

```python
# Example 6: Multiple conditions with AND
print("\n--- Example 6: WHERE with AND ---")
query = "SELECT name, age, gpa FROM students WHERE age >= 17 AND gpa > 3.0"
result = pd.read_sql_query(query, conn)
print(result)
```

`AND` means both conditions must be true.
This finds students who are 17+ years old AND have a GPA above 3.0

```python
# Example 7: Multiple conditions with OR
print("\n--- Example 7: WHERE with OR ---")
query = "SELECT name, grade_level FROM students WHERE grade_level = 9 OR grade_level = 12"
result = pd.read_sql_query(query, conn)
print(result)
```

`OR` means at least one condition must be true.
This finds students in either grade 9 OR grade 12.

# PART 4: ORDER BY - Sorting Results (8 minutes)

`ORDER BY` sorts your results.
Syntax: `SELECT` columns `FROM` table `ORDER BY` column `[ASC|DESC]`
`ASC` = ascending (smallest to largest) - this is the default
`DESC` = descending (largest to smallest)

```python
# Example 8: Sort by one column
print("\n--- Example 8: ORDER BY ascending ---")
query = "SELECT name, gpa FROM students ORDER BY gpa ASC"
result = pd.read_sql_query(query, conn)
print(result)
"""
This sorts students from lowest to highest GPA.
""";
```

```python
# Example 9: Sort in descending order
print("\n--- Example 9: ORDER BY descending ---")
query = "SELECT name, gpa FROM students ORDER BY gpa DESC"
result = pd.read_sql_query(query, conn)
print(result)

"""
This sorts students from highest to lowest GPA.
""";
```

```python
# Example 10: Sort by multiple columns
print("\n--- Example 10: ORDER BY multiple columns ---")
query = "SELECT name, grade_level, gpa FROM students ORDER BY grade_level DESC, gpa DESC"
result = pd.read_sql_query(query, conn)
print(result)

"""
This first sorts by grade_level (highest first), 
then within each grade, sorts by GPA (highest first).
""";
```

# PART 5: Aggregate Functions (COUNT, AVG, MIN, MAX, SUM) (10 minutes)
Aggregate functions perform calculations on multiple rows.
Common functions:
- `COUNT`: counts rows
- `AVG`: calculates average
- `MIN`: finds minimum value
- `MAX`: finds maximum value
- `SUM`: adds up values

```python
# Example 11: Count all students
print("\n--- Example 11: COUNT ---")
query = "SELECT COUNT(*) AS total_students FROM students"
result = pd.read_sql_query(query, conn)
print(result)

"""
COUNT(*) counts all rows in the table.
AS total_students gives the result a readable name.
""";
```

```python
# Example 12: Average GPA
print("\n--- Example 12: AVG (average) ---")
query = "SELECT AVG(gpa) AS average_gpa FROM students"
result = pd.read_sql_query(query, conn)
print(result)

"""
AVG calculates the average of all GPA values.
""";
```

```python
# Example 13: Multiple aggregate functions
print("\n--- Example 13: Multiple aggregates ---")
query = """
SELECT 
    COUNT(*) AS total_students,
    AVG(gpa) AS avg_gpa,
    MIN(gpa) AS lowest_gpa,
    MAX(gpa) AS highest_gpa
FROM students
"""
result = pd.read_sql_query(query, conn)
print(result)

"""
You can use multiple aggregate functions at once!
This gives us a summary of the entire table.
""";
```

# PART 6: GROUP BY - Grouping Data
`GROUP BY`groups rows with the same values and lets you aggregate each group.
Syntax: `SELECT` columns, aggregate_function `FROM` table `GROUP BY` column

```python
# Example 14: Count students by grade level
print("\n--- Example 14: GROUP BY with COUNT ---")
query = """
SELECT grade_level, COUNT(*) AS student_count
FROM students
GROUP BY grade_level
ORDER BY grade_level
"""
result = pd.read_sql_query(query, conn)
print(result)

"""
This groups students by grade_level and counts how many are in each grade.
ORDER BY makes the output easier to read.
"""
```

```python
# Example 15: Average GPA by grade level
print("\n--- Example 15: GROUP BY with AVG ---")
query = """
SELECT grade_level, AVG(gpa) AS average_gpa
FROM students
GROUP BY grade_level
ORDER BY grade_level
"""
result = pd.read_sql_query(query, conn)
print(result)

"""
This shows the average GPA for each grade level.
"""
```

```python
# Example 16: Multiple aggregates with GROUP BY
print("\n--- Example 16: GROUP BY with multiple aggregates ---")
query = """
SELECT 
    grade_level,
    COUNT(*) AS num_students,
    AVG(gpa) AS avg_gpa,
    MIN(gpa) AS min_gpa,
    MAX(gpa) AS max_gpa
FROM students
GROUP BY grade_level
ORDER BY grade_level
"""
result = pd.read_sql_query(query, conn)
print(result)

"""
This gives comprehensive statistics for each grade level.
"""
```

# PART 7: HAVING - Filtering Grouped Data
`HAVING` is like `WHERE`, but it filters `AFTER` grouping.
Use `WHERE` to filter individual rows.
Use `HAVING` to filter groups.

```python
# Example 17: HAVING with GROUP BY
print("\n--- Example 17: HAVING clause ---")
query = """
SELECT grade_level, AVG(gpa) AS avg_gpa
FROM students
GROUP BY grade_level
HAVING AVG(gpa) > 3.5
"""
result = pd.read_sql_query(query, conn)
print(result)

"""
This shows only grade levels where the average GPA is above 3.5.
HAVING filters the grouped results.
""";
```

```python
# Example 18: WHERE and HAVING together
print("\n--- Example 18: WHERE and HAVING together ---")
query = """
SELECT grade_level, COUNT(*) AS num_students, AVG(gpa) AS avg_gpa
FROM students
WHERE age >= 16
GROUP BY grade_level
HAVING COUNT(*) >= 2
"""
result = pd.read_sql_query(query, conn)
print(result)

"""
WHERE filters individual students (only those 16+)
Then GROUP BY groups them by grade level
Then HAVING filters groups (only grades with 2+ students)
""";
```
