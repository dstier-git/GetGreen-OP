# Intro
Here you will put what you learned about prompts into practice. We provided a sample list of data fields. These will change in the future, but your structure and your prompt is what matters.

```python
# data names, values, and mappings
data_list = ["user id", "consistency rating", "recent action category", "favorite action category", "our recommended action"]
data_vals = [1, 2, "recycle", "commuting", "recycle cans"] # random stuff
data_dict = {name: val for name, val in zip(data_list, data_vals)}
data_dict
```

# Task 1: Parse

Your first task is to **fill in the parse function** with code that uses a loop and f-strings to build the data into a string.

```python
# Use an f-string and a loop to build a prompt for the model

# Simplified example output: "the data includes this name with this value, and this name with this value, etc."
def parse(data_list, data_dict):
    parsed_string = ""
    
    # loops thru data names
    for data_name in data_list:
        # access data with:    data_dict[data_name]     -- returns the value for given data name
        # replace f"..." with your custom f-string parsing the data together
        value = data_dict[data_name]
        parsed_string = parsed_string + " " + f"{data_name}: {value},"
        
    return parsed_string
```

This cell tests your code:


```python
print(f"""Inputs:
        Data names: {data_list}
        Data values: {data_vals}
Your output:
        {parse(data_list, data_dict)}
        """)
```

# Task 2: Build Prompt
Your second task is to fill in the prompt function with code. Please be creative and thoughtful with this! We'll the best response for the actual project.

```python
# user_query is a string of the users question
def prompt(context, user_query, data_list, data_dict):
    # use the user_query and the output from the parse function for a readable, full-sentences prompt for the model that returns
    # a string combining the user's message and the context for answering
    
    parse_output = parse(data_list, data_dict)
    
    full_prompt_string =(
        f"Here is the user data: {parse_output}. "
        f"The user asked: '{user_query}'. "
        f"{context}"
    ) # fill this out
    return full_prompt_string
```
